#IMPORT LIBARIES AND DATASETS

```python
# import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Predicting Cardiovascular disease

```python

df = pd.read_csv("cardio_train.csv", sep=";")
```

```python
df.head()
```

```python
df.tail()
```

#PERFORM EXPLORATORY DATA ANALYSIS

```python

df = df.drop(columns = 'id')
```

```python
# since the age is given in days, we convert it into years
df['age'] = df['age']/365
```

```python
df.head()
```

```python
# Statistical summary of the dataframe
df.describe()
```

```python
df.hist(bins = 30, figsize = (10,10), color = 'r')
plt.show()
```

```python
# get the correlation matrix
corr_matrix = df.corr()
corr_matrix
```

```python
# plotting the correlation matrix
plt.figure(figsize = (16,16))
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

#CREATE TRAINING AND TESTING DATASET

```python
# split the dataframe into target and features
y = df['cardio']
X = df.drop(columns =['cardio'])
```

```python
X.shape
```

```python
y.shape
```

```python
#spliting the data in to test and train sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

#TRAIN AN XG-BOOST CLASSIFIER IN SK-LEARN

```python
from xgboost import XGBClassifier
```

```python
# Train an XGBoost classifier model

xgb_classifier = XGBClassifier(objective ='binary:logistic', eval_metric = 'error', learning_rate = 0.1, max_depth = 10, n_estimators = 10)
xgb_classifier.fit(X_train, y_train)
```

#TEST XGBOOST CLASSIFIER TO PERFORM INFERENCE

```python
# predict the score of the trained model using the testing dataset
result = xgb_classifier.score(X_test, y_test)
print("Accuracy:{}".format(result))
```

```python
# make predictions on the test data
y_predict = xgb_classifier.predict(X_test)
y_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
# precision is the ratio of TP/(TP+FP)
# recall is the ratio of TP/(TP+FN)
# F-beta score can be interpreted as a weighted harmonic mean of the precision and recall
# where an F-beta score reaches its best value at 1 and worst score at 0.
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, fmt = 'd', annot = True)
```

# Diabetes Prediction

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# You have to include the full link to the csv file containing your dataset
df = pd.read_csv('diabetes.csv')
```

```python
df.info()
```

```python
# Plot Histogram
df.hist(bins = 30, figsize = (20,20), color = 'b');
```

```python
# Plot the correlation matrix
correlations = df.corr()
f, ax = plt.subplots(figsize = (20, 20))
sns.heatmap(correlations, annot = True);
```

```python
y = df['Outcome']
y
```

```python
X = df.drop(['Outcome'], axis = 1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Train an XGBoost classifier model

xgb_classifier = XGBClassifier(objective ='binary:logistic', eval_metric = 'error', learning_rate = 0.1, max_depth = 1, n_estimators = 10, use_label_encoder=False)
xgb_classifier.fit(X_train, y_train)
```

```python

result = xgb_classifier.score(X_test, y_test)
print("Accuracy :{}".format(result))
```

```python
# make predictions on the test data
y_predict = xgb_classifier.predict(X_test)
y_predict
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, fmt = 'd', annot = True)
```
